```python
# Pstance数据预处理
# Data Loading
def load_data(filename):

    filename = [filename]
    concat_text = pd.DataFrame()
    raw_text = pd.read_csv(filename[0],usecols=[0], encoding='ISO-8859-1')
    raw_label = pd.read_csv(filename[0],usecols=[2], encoding='ISO-8859-1')
    raw_target = pd.read_csv(filename[0],usecols=[1], encoding='ISO-8859-1')
    label = pd.DataFrame.replace(raw_label,['FAVOR','NONE','AGAINST'], [1,2,0])
    concat_text = pd.concat([raw_text, label, raw_target], axis=1)
    concat_text = concat_text[concat_text.Stance != 2]
    
    return(concat_text)

# Data Cleaning
def data_clean(strings, norm_dict):
    
    p.set_options(p.OPT.URL,p.OPT.EMOJI,p.OPT.RESERVED)
    clean_data = p.clean(strings)  # using lib to clean URL, emoji...
    clean_data = re.sub(r"#SemST", "", clean_data)
    clean_data = re.findall(r"[A-Za-z#@]+|[,.!?&/\<>=$]|[0-9]+",clean_data)
    clean_data = [[x.lower()] for x in clean_data]
    
    for i in range(len(clean_data)):
        if clean_data[i][0] in norm_dict.keys():
            clean_data[i][0] = norm_dict[clean_data[i][0]]
            continue
        if clean_data[i][0].startswith("#") or clean_data[i][0].startswith("@"):
            clean_data[i] = wordninja.split(clean_data[i][0]) # split compound hashtags
    clean_data = [j for i in clean_data for j in i]

    return clean_data

# Clean All Data
def clean_all(filename, norm_dict):
    
    concat_text = load_data(filename)
    raw_data = concat_text['Tweet'].values.tolist() 
    label = concat_text['Stance'].values.tolist()
    x_target = concat_text['Target'].values.tolist()
    clean_data = [None for _ in range(len(raw_data))]
    
    for i in range(len(raw_data)):
        clean_data[i] = data_clean(raw_data[i], norm_dict)
        x_target[i] = data_clean(x_target[i], norm_dict)
    
    return clean_data,label,x_target

```

```python
def read_and_clean(target_index):
    target_word_pair=['bernie','biden','trump']
    homepath = 'D:/study/code/TexAtt/PStance'
    filename1 = homepath + '/raw_train_' + target_word_pair[target_index]+'.csv'
    filename2 = homepath + '/raw_val_' + target_word_pair[target_index]+'.csv'
    filename3 = homepath + '/raw_test_' + target_word_pair[target_index]+'.csv'
    
    x_train,y_train,x_train_target = clean_all(filename1, normalization_dict)
    x_val,y_val,x_val_target = clean_all(filename2, normalization_dict)
    x_test,y_test,x_test_target = clean_all(filename3, normalization_dict)
    
    def process_sublists(sublists):
        processed = []
        # 对于每一个子列表，执行join_words函数，并将结果添加到新的列表中
        for sublist in sublists:
            processed.append(join_words(sublist))
        return processed

    x_train_ = process_sublists(x_train)
    x_train_target_ = process_sublists(x_train_target)
    
    x_val_ = process_sublists(x_val)
    x_val_target_ = process_sublists(x_val_target)

    x_test_ = process_sublists(x_test)
    x_test_target_ = process_sublists(x_test_target)

    # 创建dataframes
    train_df = pd.DataFrame({'text_a': x_train_, 'Target': x_train_target_, 'label': y_train})
    val_df = pd.DataFrame({'text_a': x_val_, 'Target': x_val_target_, 'label': y_val})
    test_df = pd.DataFrame({'text_a': x_test_, 'Target': x_test_target_, 'label': y_test})

    return train_df, val_df, test_df
    
def join_words(words_list):
    # 将所有单词拼接起来，用空格分隔
    sentence = ' '.join(words_list)
    # 将空格加标点符号替换为没有空格的标点符号
    sentence = sentence.replace(' .', '.').replace(' ,', ',').replace(' ?', '?').replace(' !', '!').replace(' ;', ';').replace(' &', '&')
    return sentence
```

```python
#wt-wt 的数据处理
def split_and_rename(df):
    # 按照给定的比例划分数据集
    train_size = 0.7
    val_test_size = 0.15  # 验证集和测试集的大小
    train, temp = train_test_split(df, test_size=1-train_size, random_state=42) # 设置random_state保证可重复性
    val, test = train_test_split(temp, test_size=val_test_size/(val_test_size*2), random_state=42) # 再次划分得到验证集和测试集

    # 重命名列
    renamed_columns = {'text': 'text_a', 'merger': 'Target', 'stance': 'label'}
    train = train.rename(columns=renamed_columns)
    val = val.rename(columns=renamed_columns)
    test = test.rename(columns=renamed_columns)

    # 替换 'label' 列中的值
    replacements = {'refute': 0, 'support': 1, 'comment': 2, 'unrelated': 3}
    train['label'] = train['label'].replace(replacements)
    val['label'] = val['label'].replace(replacements)
    test['label'] = test['label'].replace(replacements)

    return train, val, test

def save_split_dataframes(group_dict, norm_dict):
    for merger, df in group_dict.items():
        # 清理dataframe
        df = clean_dataframe(df, norm_dict)

        # 使用之前定义的函数来划分和重命名dataframe
        train, val, test = split_and_rename(df)

        # 创建一个目录来存储结果的CSV文件
        os.makedirs(merger, exist_ok=True)

        # 将结果的dataframe保存为CSV文件
        train.to_csv(f'{merger}/train.csv', index=False)
        val.to_csv(f'{merger}/dev.csv', index=False)
        test.to_csv(f'{merger}/test.csv', index=False)

def data_clean(strings, norm_dict):
    
    p.set_options(p.OPT.URL, p.OPT.EMOJI, p.OPT.RESERVED)
    clean_data = p.clean(strings)  # 使用库清理URL、表情符号等
    clean_data = re.sub(r"#SemST", "", clean_data)
    clean_data = re.findall(r"[A-Za-z#@]+|[,.!?&/\<>=$]|[0-9]+",clean_data)
    clean_data = [[x.lower()] for x in clean_data]
    
    for i in range(len(clean_data)):
        if clean_data[i][0] in norm_dict.keys():
            clean_data[i][0] = norm_dict[clean_data[i][0]]
            continue
        if clean_data[i][0].startswith("#") or clean_data[i][0].startswith("@"):
            clean_data[i] = wordninja.split(clean_data[i][0]) # 分割复合型hashtag
    clean_data = [j for i in clean_data for j in i]

    return ' '.join(clean_data)  # 返回字符串，而不是词列表

def clean_dataframe(df, norm_dict):
    df['text'] = df['text'].apply(lambda x: data_clean(x, norm_dict))
    return df
```
